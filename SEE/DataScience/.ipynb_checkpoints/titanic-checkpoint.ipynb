```python
import numpy as np
import pandas as pd
```

```python
df3=pd.read_csv('titanictrain.csv')
```

```python
df3.Survived=df3.Survived.map({0:'D',1:'S'})
```

```python
df3.drop(['Parch','Fare'],axis=1,)
```
