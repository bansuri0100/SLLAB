```python
import numpy as np
import pandas as pd
```

```python
titanic_dt=pd.read_csv('titanictrain.csv')
```

```python
titanic_dt.head()
```

```python
titanic_dt ['Survived'] = titanic_dt ['Survived'].map({ 0: 'Died', 1:'Survived'})
```

```python
titanic_dt.drop(['Parch','PassengerId','Name','Ticket','Embarked'],axis=1, inplace=False)
```

```python
titanic_dt['Embarked'] = titanic_dt['Embarked'].fillna('S')
```
